```python
import json
import csv
import tweepy
import re
import os

def twitter_hashtag(api_key, api_key_secret, access_token, access_token_secret, hashtag):
    
    auth = tweepy.OAuthHandler(api_key, api_key_secret)
    auth.set_access_token(access_token, access_token_secret)

    api = tweepy.API(auth)
    
    file_name = '_'.join(re.findall(r"#(\w+)", hashtag))

    with open('%s.csv' % (file_name), 'w') as file:

        w = csv.writer(file)

        w.writerow(['tweet_text'])

        for tweet in tweepy.Cursor(api.search, q=hashtag+' -filter:retweets', lang="en", tweet_mode='extended').items(100):
            w.writerow([tweet.full_text.replace('\n',' ').encode('utf-8')])
            
api_key = input('Please give twitter api key')
api_key_secret = input('Please give twitter api key secret')
access_token = input('Please give twitter access token')
access_token_secret = input('Please give twitter access token secret')
    
hashtag = input('enter the hashtag')

twitter_hashtag(api_key, api_key_secret, access_token, access_token_secret, hashtag)
```

```python
import pandas as pd
df = pd.read_csv("CovidVaccine.csv")
df.head()
```

```python
!pip install nltk
import nltk
#Remove noise
df['tweet_text'] = df['tweet_text'].str.replace('[^\w\s]','')
#df['tweet_text'].head()

#Remove stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['tweet_text'] = df['tweet_text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df['tweet_text']

#Remove  numbers
df['tweet_text'] = df['tweet_text'].apply(lambda x: " ".join(x for x in x.split() if not x.isdigit()))
df['tweet_text']

#lowercase
df['tweet_text'] = df['tweet_text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['tweet_text']

#stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['tweet_text'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))

#lemmatization
!pip install -U textblob
from textblob import Word
import nltk
nltk.download('wordnet')

df['tweet_text'] = df['tweet_text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df.head()
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk import word_tokenize, pos_tag
from functools import partial
tags_tokens = lambda x: pos_tag(word_tokenize(x))
from collections import Counter
from itertools import chain

df['pos_sentences'] = df['tweet_text'].apply(tags_tokens)
df.head()
```

```python
tokens, tags = zip(*chain(*df['pos_sentences'].tolist()))
print(tokens)
print(tags)
p_tags = sorted(set(list(zip(*chain(*df['pos_sentences'])))[1]))
```

```python
def add_all_pos(counter, keys):
    print("counter:",counter)
    for i in keys:
        counter[i] = counter.get(i, 0)
    return counter
```

```python
df['tags_counts'] = df['pos_sentences'].apply(lambda x: Counter(list(zip(*x))[1]))
#df['tags_counts'] #These are tags of pos which are in the tweets
df['total_counts'] = df['tags_counts'].apply(lambda x: add_all_pos(x, p_tags))
#df['total_counts'] #These are zero pos tags plus other counted tags
df['vector'] = df['total_counts'].apply(lambda x: [count for tag, count in sorted(x.most_common())])
#df['vector']
```

```python
#adding vector to respective pos_sentences

df['vector'] = df['pos_sentences'].apply(lambda x:
    [count for tag, count in sorted(
        add_all_pos(
            Counter(list(zip(*x))[1]), 
                    p_tags).most_common()
         )
    ]
)
```

```python
df_final = pd.DataFrame(df['vector'].tolist())
df_final.columns = p_tags
df_final['tweets']=df['tweet_text']
```
