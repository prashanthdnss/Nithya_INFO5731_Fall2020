```python
# Write your code here
import nltk
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
nltk.download('stopwords')
stop=stopwords.words('english')
#print(stop)
file = open('C:/Users/prash/OneDrive/Desktop/INFO 5731 Computational Methods/In Class Exercieses/Exercise 4/raw_data.txt', "r", encoding="utf8")
data = file.read()
words = data.split()
print('Number of words in text file :', len(words))
sentences = data.split('.')
print('Number of sentences in text file :', len(sentences))

print('Number of characters in text file :', len(data))

average = sum(len(word) for word in words) / len(words)
print('Number of characters in text file :', average)

stop_words = 0
count_special = 0
numerical =0
upper=0

for w in words:
    if w in stop:
        stop_words=stop_words+1
    for i in range(len(w)):
        if(w[i].isalpha()):
            continue
        elif(w[i].isdigit()):
            numerical=numerical+1
        else:
            count_special = count_special + 1
            
    if w.isupper():
        upper=upper+1
    
print('Number of stopwords in text file :', stop_words)

print('Number of special characters in text file :', count_special)

print('Number of numericals in text file :', numerical)

print('Number of uppercase in text file :', upper)

```
